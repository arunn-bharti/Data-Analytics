# Subplots (can't handle canvas)

```python
plt.subplot(2,2,1) #(no of row, no of col, which plot you choose) i choose 2,2 so i can make 4 subplots only
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,x)
plt.subplot(2,2,3)
plt.plot(y,x**4)
plt.subplot(2,2,4)
plt.plot(x,x//4)
```

# Object oriented way (to handle canvas)

```python
fig=plt.figure()
axis1=fig.add_axes([.1,.1,.8,.8]) #here [Persentage from left, then bottom, its for width, its for height]
axis2=fig.add_axes([.4,.4,.5,.5])
axis3=fig.add_axes([.2,.2,.5,.5])
axis1.plot(x,y)
axis2.plot(x,y)
axis3.plot(x,y)
```

```python
fig=plt.figure(figsize=(20,10),dpi=200) #figsiz=(height, width) in inches and dpi for high resolution 
ax1=fig.add_axes([.1,.1,.8,.8])
ax1.plot(x,y)
```

# Distribution Plots (Start of seaborn)

```python
# ploting on fields which are not categorised, like in salary,age,gender,  here age n salary is distributive and gender is categorised
```

```python
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
#import dataframe from seaborn library
df1=sb.load_dataset('tips')
df1
```

```python
df1['size'].unique()   # to get unique values from particuler  field
```

```python
#1. distplot/histplot (newly use histplot)
plt.subplot(1,2,1)
sb.histplot(df1['tip'],bins=10,kde=True)  #used to represents one field data in bins by default its 20 and bins means no of boxes / and kde(kernel density estimate) for that line
plt.subplot(1,2,2)
sb.histplot(df1['total_bill'],bins=10,kde=True)
```

```python
#2. Joinplot
sb.jointplot(x='total_bill', y='tip', data=df1, kind='scatter')  #here kind tells the what kinda graph we need, by default scatter
```

```python
sb.jointplot(x='total_bill', y='tip', data=df1, kind='reg')
```

```python
sb.jointplot(x='total_bill', y='tip', data=df1, kind='kde')
```

```python
sb.jointplot(x='total_bill', y='tip', data=df1, kind='resid')
```

```python
sb.pairplot(data=df1,hue='sex',palette='rainbow',kind='hist')
```

```python
sb.rugplot(df1) #not imp
```

# Categorized Plots

```python
sb.countplot(x=df1['sex'],hue=df1['smoker'],hue_order=['No','Yes']) #countplot takess only one position(x or y), hue for group data, and hue_order for sequence to display grouped data
```

```python
df1
```

```python
sb.barplot(y=df1['tip'],x=df1['sex']) #have 2 aixs and one axis should have distributive field and another categorised
```

bar plot by default estimator= avg(mean) for distributed field. but by providing estimator we can change it to sum, min, max etc

```python
sb.barplot(y=df1['tip'],x=df1['sex'],estimator=np.sum)
```

```python
sb.boxplot(x='tip',y='day',data=df1) #that circles are outliers and to comparison bt categorised and distributive fields
```
