# While loop

```python
x=0
while x<=10:
    print(x)
    x+=1
    
```

# For loop

```python
x=[1,2,3,4]
for a in x:
    print(a)
```

# Nested for loops

```python
x=[[1,2,3],['a','b','c']]
for a in x:
    for b in a:
        print(b,end="")
    print()
```

```python
for i in range(1, 6):         # Outer loop for rows
    for j in range(1, i + 1): # Inner loop for numbers in each row
        print(j, end=" ")
    print()                   # Move to the next line
```

```python
for i in range(5,0,-1):
    for j in range(i):
        print("*",end="")
    print()
```

```python
for i in range(6,1,-1):
    for j in range(1,i+1):
        print(j,end="")
    print()
```

```python
rows=int(input("enter # of rows : "))
columns=int(input("enter # of columns : "))
symbol=input("enter symbol to print : ")

for x in range(rows):
    for y in range(columns):
        print(symbol,end="")
    print()
```

```python
n=int(input("enter a number to reverse : "))
rev_num=0
while n%10!=0:
    reminder=n%10
    rev_num=rev_num*10+reminder
    n=n//10
print("reverse number is : ",rev_num)
```

# Break and Continue 

```python
txt="hey! how are you"
for i in txt:
    if i=="!":
        break
    print(i,end="")
```

```python
txt="hey! how are you"
for i in txt:
    if i=="!":
        continue
    print(i,end="")
```
