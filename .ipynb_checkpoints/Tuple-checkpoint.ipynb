# Tuple

```python
chars='a','b','c','d' #tuble may or may not be in round braces. so this is also a tuple.
type(chars)
```

```python
Tuple1=(1,2,3,4)
print(Tuple1)
```

```python
type(Tuple1)
```

```python
print(Tuple1[2])
```

```python
Tuple1[3]=10 
print(Tuple1)
#tupple is immutable. we can't change it, remove it.
```

```python
z=("raj","rahul","dipak",1,2,3)
```

# Slicing

```python
print(z[:5])
```

```python
print(z[::3])
```

# Concatenation

```python
num=(1,2,3,4)
str=("jaipur","mumbai","surat")
```

```python
print(num+str)
```

# Nesting

```python
nest=(num+str)
print(nest)
```

# Repetition

```python
tup=("Python",)*5
print(tup)
```

```python
tup=("Python",)
print(tup*10)
```

```python
print(tup)
```

# Unpacking

```python
vowels="aeiou"
tuple(vowels)
```

```python
numbers=1,2,3,4
a,b,c,d=numbers
print(a,b,c,d)
```

```python
a,*b,c=numbers
print(a,b,c)
```

# Delete a tuple

```python
new_tup=(12,15,19)
print(new_tup)
```

```python
del new_tup
print(new_tup) #it shows an error bcs tuple is already deleted.
```

# Converting List to Tuple

```python
lst=[2,3,"hey"]
type(lst)
```

```python
list_tupled=tuple(lst)
print(list_tupled)
```

```python
type(list_tupled)
```

```python
lst=[(1,2,3),("a","b","c")]
print(lst)
```

```python
lst.append(("hello", "hey")) #list can be appended.
lst
```
