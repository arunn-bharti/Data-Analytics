```python
n=int(input("enter a number : "))
if n>0:
    print("positive number")
elif n<0:
    print("negative number")
else:
    print("it's zero")
```

```python
import math
#area calculator
ch=input("enter your choice (rectangle, triangle, circle or square) : ")
if ch=="rectangle":
    l=float(input("enter length of rectange : "))
    w=float(input("enter width of rectange : "))
    print(f"area of rectangle is {l*w}")
elif ch=="triangle":
    b=float(input("enter base of triangle : "))
    h=float(input("enter height of triangle : "))
    print(f"area of triangle is {1/2*b*h}")
elif ch=="circle":
    r=float(input("enter radius of circle : "))
    print(f"area of circle is {math.pi*r*r}")
elif ch=="square":
    a=float(input("enter one side lenth of square : "))
    print(f"area of square is {a*a}")
else:
    print("invalid choice!")
```

```python
ch=input("enter any character : ")

if ch=='a'or ch=='e' or ch=='i' or ch=='o' or ch=='u' or ch=='A' or ch=='E' or ch=='I' or ch=='O' or ch=='U':
    print("its a vowel")
else:
    print("it's a consonants")
```

```python
#check number length

n=input("enter a number : ")
ln=len(n)
if ln==1:
    print("it's a single digit number")
elif ln==2:
    print("it's a two digit number")
elif ln==3:
    print("it's a three digit number")
elif ln==4:
    print("it's a four digit number")
else:
    print("it's a complex number number!")
```

```python
#check number length-method 2

n=int(input("enter a number : "))
if n>0 and n<=9:
    print("it's a single digit number")
elif n>9 and n<=99:
    print("it's a two digit number")
elif n>99 and n<=999:
    print("it's a three digit number")
elif n>999 and n<=9999:
    print("it's a four digit number")
else:
    print("it's a complex number number!")
```
