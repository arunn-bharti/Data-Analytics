# String Function Part 1

```python
s="You think i am hiding in shadow, but you don't know I AM THE SHADOW"
name="Arun"
print(type(s))
print(len(s))
print(s.count("i"))
print(s.upper())
print(s.lower())
print(s.title())
print(s.capitalize())
print(s.casefold())
print(s.index("I"))
print(s.find("I"))
print(s.center(100,"*"))
print(f"hello {name}")
```

# String Function Part 2

```python
a="Arun"
b="23"
c="3.14"
d="aru123"
u="ARUN"
l="arun"
t="The Batman"
g=" "


print(d.isalnum())
print(a.isalpha())
print(b.isnumeric())
print(b.isdecimal())
print(l.islower())
print(u.isupper())
print(g.isspace())
print(t.istitle())
print(b.isdigit())
```

```python
x="arun"
print(x.ljust(20),"how are you")
print(x.ljust(20,"-"),"how are you")
print(x.rjust(20),"how are you")
print(x.rjust(20,"-"),"how are you")
```

```python
print(a.replace("harry","james"))
```

```python
print(a.rindex("t")) #last accurence
print(a.rindex("g",10,25))
print(a.index("t")) #first accurence
print(a.rfind("o")) #last accurence
print(a.find("x")) #find prints -1 if given string is not available in string.
print(a.find("o")) #first accurence
```
