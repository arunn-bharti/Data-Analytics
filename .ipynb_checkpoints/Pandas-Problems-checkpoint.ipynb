```python
import numpy as np

df_final['ord_size_tag'] = df_final['quantity'].apply(
    lambda x: 'bulk' if x > 4 else ('standard' if x >= 2 else 'single')
)
```

```python
df_final.pivot_table(index='category',columns='store_type',values='quantity',aggfunc='sum').style.background_gradient(axis=0)
```

```python
# 1. Get total revenue per product and sort it
prod_rev = df_final.groupby('product_name')['revenue'].sum().sort_values(ascending=False)

# 2. Calculate the cumulative sum of revenue
cumulative_revenue = prod_rev.cumsum()

# 3. Calculate the total revenue of everything
total_rev_all = df_final['revenue'].sum()

# 4. Find how many products make up 50% of total revenue
half_revenue_mark = total_rev_all / 2
products_contributing_50 = cumulative_revenue[cumulative_revenue <= half_revenue_mark]

print(f"Total Products: {len(prod_rev)}")
print(f"Number of products contributing to 50% of revenue: {len(products_contributing_50)}")
print("\nTop 5 Revenue Drivers:")
print(prod_rev.head(5))
```

```python
# 1. Convert to datetime (just in case it isn't already)
df_final['order_date'] = pd.to_datetime(df_final['order_date'])

# 2. Create the 'day_of_week' column (0=Mon, 6=Sun)
df_final['day_of_week'] = df_final['order_date'].dt.dayofweek

# 3. Create 'is_weekend' using np.where
# 5 is Saturday, 6 is Sunday
df_final['is_weekend'] = np.where(df_final['day_of_week'] >= 5, 'Weekend', 'Weekday')

# 4. Group by your new column and find average revenue
weekend_analysis = df_final.groupby('is_weekend')['revenue'].mean()

print(weekend_analysis)
```
