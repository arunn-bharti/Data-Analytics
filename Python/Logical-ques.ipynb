```python
n=int(input("Enter number what length of fibonacci series youu want"))
a=0
b=1
print(a,b,end=" ")
for i in range(1,n):
    next=a+b
    print(next,end=" ")
    a=b
    b=next
    
```

```python
n=int(input("enter a number : "))
isprime=True

if n==1:
    isprime=False
else: 
    for i in range(2,n):
        if n%i==0:
            isprime=False
            break
if isprime:
    print("it is prime number")
else:
    print("it is not a prime number")
```

```python
n=int(input("enter num "))
c=n
rev=0
while n>0:
    rem=n%10
    rev=rev*10+rem
    n=n//10
print(rev)

print("polindrome") if c==rev else print("not polindrome")
```
