# if statement

```python
a=10
if a>5:
    print("a greater than 5")
```

# if-else statement

```python
a=27
if a%2==0:
    print("it's an even number")
else:
    print("it's an odd number")
```

# if-elif-else statement

```python
char=input("enter a character : ")
if char=='a':
    print("it's a vowel")
elif char=='e':
    print("it's a vowel")
elif char=='i':
    print("it's a vowel")
elif char=='o':
    print("it's a vowel")
elif char=='u':
    print("it's a vowel")
else:
    print("it's a consonant")
```

# Nested if

```python
a=int(input("enter an integer"))
if a>0:
    print("it's greater than zero")
    if a%2==0:
        print("it's even")
    else:
        print("it's odd")
else:
    print("it's less than zero")
```
