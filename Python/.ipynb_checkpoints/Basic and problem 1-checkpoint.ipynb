```python
#swaping 

a=10
b=20
print(f"before swapping a={a} and b={b}")
a=a+b
b=a-b
a=a-b
print(f"after swapping a={a} and b={b}")
```

```python
#program to take user input for student id card
from datetime import datetime
name=input("enter student name : ")
dob=input("enter student dob (in YYYY-MM-DD format) : ")
add=input("enter address : ")

f_dob=datetime.strptime(dob,"%Y-%m-%d").date()
print("\nstudent id card")
print(f"\nname : {name}")
print(f"date of birth : {f_dob}")
print(f"address : {add}")
```
