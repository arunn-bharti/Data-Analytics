```python
#billing system

while True:
    name=input("Enter customer name : ")
    total=0
    qnt=0
    while True:
        qnt=int(input("Enter quantity : "))
        price=int(input("Enter price : "))
        total+=qnt*price
        new_item=input("do you want to buy another item : ")
        if new_item=="no":
            break
    print(f"Customer name : {name}")
    print(f"quantity : {qnt}")
    print(f"total amount to pay : {total}")

    new_cust=input("new customer?")
    if new_cust=="no":
        break
```
